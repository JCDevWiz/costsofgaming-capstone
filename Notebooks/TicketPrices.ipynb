```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
tickets = pd.read_csv('../Data/movie-ticket_prices.csv')
tickets.head()
```

```python
tickets = tickets.dropna()
```

```python
tickets
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
sns.regplot(
    x=tickets["Year"],
    y=tickets["Price"],
    line_kws={"color": "red", "linewidth": 1.5},
    ax=ax
)
```

```python
tickets['Inflation'] = tickets['Price'].pct_change() * 100
```

```python
tickets
```

```python
base_price = tickets.loc[tickets['Year'] == 1976, 'Price'].iloc[0]
```

```python
base_price
```

```python
tickets['Total_Inflation'] = ((tickets['Price'] / base_price) - 1) * 100
```

```python
tickets
```
